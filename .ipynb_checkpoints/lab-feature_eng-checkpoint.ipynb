```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsRegressor

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

**Check for data types**

```python
column_summary_spaceship = pd.DataFrame({
    'Column Name': spaceship.columns,
    'Data Type': spaceship.dtypes.values,
    'Percentage Null': spaceship.isnull().mean().values * 100,
    'Unique Values': spaceship.nunique().values
})

print(column_summary_spaceship)
```

**Check for missing values**

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship = spaceship.dropna()
```

```python
spaceship.shape
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
print(list(spaceship['Cabin'].unique()))
```

```python
spaceship['Cabin'] = spaceship['Cabin'].apply(lambda x: x[0] if pd.notna(x) else None)
```

```python
spaceship['Cabin'].unique()
```

- Drop PassengerId and Name

```python
spaceship = spaceship.drop(columns=['PassengerId', 'Name'])
```

- For non-numerical columns, do dummies.

```python
non_numeric_spaceship = spaceship.select_dtypes(exclude=['number'])

column_summary_spaceship = pd.DataFrame({
    'Column Name': non_numeric_spaceship.columns,
    'Data Type': non_numeric_spaceship.dtypes.values,
    'Percentage Null': non_numeric_spaceship.isnull().mean().values * 100,
    'Unique Values': non_numeric_spaceship.nunique().values
})

print(column_summary_spaceship)
```

```python
non_numeric_cols = spaceship.select_dtypes(exclude=['number'])

df_dummies = pd.get_dummies(non_numeric_cols)

spaceship_with_dummies = pd.concat([spaceship.drop(non_numeric_cols.columns, axis=1), df_dummies], axis=1)


pd.set_option('display.max_columns', None)

spaceship_with_dummies.head()
```

**Perform Train Test Split**

```python
target = spaceship_with_dummies["Transported"]

features = spaceship_with_dummies.drop("Transported", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

```python
knn = KNeighborsRegressor(n_neighbors=10)
```

```python
knn.fit(X_train_norm, y_train)
```

- Evaluate your model's performance. Comment it

```python
knn.score(X_test_norm, y_test)
```
